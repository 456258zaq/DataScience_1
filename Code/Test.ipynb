```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"../Data/train.csv",index_col = "Id")
df.head()
```

```python
#check na value
print("Na Value Columns:",df.columns[df.isna().sum() > 0])
NA_col = df.columns[df.isna().sum() > 0]
print("NA distribution : \n",df.loc[:,NA_col].isna().sum())
```

```python
def na_value_percent(dataframe,column):
    df = dataframe
    na_count = df[column].isna().sum()
    print("columns : {}, na count : {}, na percent : {:.2f}".format(column,na_count,na_count/len(df)*100))
```

```python
for i in NA_col:
    na_value_percent(df,i)
```

- NA Count over 40%, need to drop column
- drop column list : Alley, PoolQC, Fence, MiscFeature

```python
df.drop(columns=["Alley","PoolQC","Fence","MiscFeature","FireplaceQu"],inplace=True)
```

```python
NA_col = NA_col.tolist()
for i in ["Alley","PoolQC","Fence","MiscFeature","FireplaceQu"]:
    NA_col.remove(i)
```

```python
NA_col
```

```python
for i in NA_col:
    print(i,": \t ",df[i].dtype)
```

```python
def make_graph(column,numpy_array):
    plt.figure(figsize=(5,5))
    plt.style.use("bmh")
    if numpy_array.dtype != "object":
        plt.rcParams['text.color'] = 'black'
        sns.distplot(numpy_array)
        plt.title(column)
    else:
        plt.rcParams['text.color'] = 'white'
        plt.pie(numpy_array.value_counts(),labels=numpy_array.value_counts().index,autopct="%.1f%%")
        plt.title(column)
```

```python
for i in NA_col:
    make_graph(i,df[i])
```

```python
df["LotFrontage"] = df["LotFrontage"].fillna(df["LotFrontage"].median())
df["MasVnrType"] = df["MasVnrType"].fillna("nothing")
df["MasVnrArea"] = df["MasVnrArea"].fillna(0)
df["BsmtQual"] = df["BsmtQual"].fillna("No Basement")
df["BsmtCond"] = df["BsmtCond"].fillna("No Basement")
df["BsmtExposure"] = df["BsmtExposure"].fillna("No Basement")
df["BsmtFinType1"] = df["BsmtFinType1"].fillna("No Basement")
df["BsmtFinType2"] = df["BsmtFinType2"].fillna("No Basement")
df["Electrical"] = df["Electrical"].fillna(df["Electrical"].mode()[0])
df["GarageType"] = df["GarageType"].fillna("No Garage")
df["GarageFinish"] = df["GarageFinish"].fillna("No Garage")
df["GarageQual"] = df["GarageQual"].fillna("No Garage")
df["GarageCond"] = df["GarageCond"].fillna("No Garage")
df["GarageYrBlt"] = df["GarageYrBlt"].fillna(0)
```
